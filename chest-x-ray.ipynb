```
import numpy as np
import pandas as pd
import sklearn.metrics as m
```

```
data=pd.read_csv("metadata.csv")
```

```
data
```

```
features=data.iloc[:,4]
```

```
features.unique()
```

```
features[features=='COVID-19']=0
```

```
from keras.utils.np_utils import to_categorical
labels=to_categorical(np.asarray(features))
```

```
labels
```

```
images=data.loc[:,['filename']]
```

```
import os
import cv2
image_read=[]
for im in images['filename']:
  print(im)
  image_path=os.path.join('/content/drive/My Drive/coronachest-xray/images/',im)
  image1=cv2.imread(image_path)
  image2=cv2.resize(image1,(200,200))
  image3=image2.astype('float32')
  image4=image3/255
  image_read.append(image4)
  
image_array=np.asarray(image_read)
  


```

```
image_array.shape
```

```
from sklearn.model_selection import train_test_split
```

```
from keras.models import Sequential
from keras.layers import Dense,Conv2D,Flatten,Activation,MaxPooling2D
```

```
model=Sequential()
model.add(Conv2D(32,(3,3),input_shape=(200,200,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(64,(3,3),input_shape=(200,200,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(64,(3,3),input_shape=(200,200,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(512,activation='relu'))
model.add(Dense(256,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(2,activation='softmax'))
```

```
model.summary()
```

```
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```
arr=[]
for i in range(1,250):
  features_train,features_test,labels_train,labels_test=train_test_split(image_array,labels,train_size=0.7,random_state=i)
  model.fit(features_train,labels_train,epochs=50,verbose=2)
  labels_predict=model.predict(features_test)
  
```

```
model.save("model.h5")
```
